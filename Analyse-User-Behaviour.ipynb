```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r'C:\Users\Videsh Loya\Desktop\Python\UserDataset.xlsx')
```

```python
df.head()
```

```python
df.set_index(df.columns[0],inplace=True)
```

```python
df.head()
```

```python
data=[]
```

```python
for i in range(100):
    x=df.values[i]
    x=x[~pd.isnull(x)]
    data.append(x)
```

```python
from markovclick.models import MarkovClickstream
m = MarkovClickstream(data)
```

```python
fig=plt.figure(figsize=(20,10), dpi=90, facecolor='w', edgecolor='k')
sns.heatmap(m.prob_matrix, xticklabels=m.pages, yticklabels=m.pages,cmap="coolwarm",annot=True)
```

```python
from markovclick.viz import visualise_markov_chain
graph = visualise_markov_chain(m)
graph
```
